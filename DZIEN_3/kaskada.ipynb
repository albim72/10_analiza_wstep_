```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#dane dla wykresu
categories = ["Start","Zysk","Strata","Zysk","Strata","Koniec"]
values = [100,30,-15,40,-10,145]
colors = ["blue" if x>=0 else "red" for x in values]
```

```python
#obliczenie pozycji słupków
start = [values[0]]
for i in range(1,len(values)):
    start.append(start[i-1]+values[i-1])
```

```python
plt.bar(categories,values,color=colors,bottom=start,edgecolor="black")

# for i in range(1,len(values)):
#   plt.plot([i-1,i],[start[i],start[i]],color="black")

for i in range(1,len(values)):
    x_start = i - 0.5
    x_end = i + 0.5
    y_start = start[i]
    y_end = start[i]
    plt.plot([x_start,x_end],[start[i],start[i]],color="black",linestyle='-',linewidth=1.5)


plt.xlabel("Kategoria")
plt.ylabel("Wartość")
plt.title("Wykres kaskadowy")
plt.grid(axis="y", linestyle="--",alpha=0.7)
plt.show()
```

```python
# prompt: wykreśl dowolny fraktal za pomocą matplotlib

import matplotlib.pyplot as plt
import numpy as np

def sierpinski(level, points):
    if level == 0:
        plt.fill(points[:, 0], points[:, 1], 'b')
        return

    p1 = points[0, :]
    p2 = points[1, :]
    p3 = points[2, :]

    mid1 = (p1 + p2) / 2
    mid2 = (p2 + p3) / 2
    mid3 = (p1 + p3) / 2

    sierpinski(level - 1, np.array([p1, mid1, mid3]))
    sierpinski(level - 1, np.array([mid1, p2, mid2]))
    sierpinski(level - 1, np.array([mid3, mid2, p3]))

points = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3) / 2]])
sierpinski(5, points)

plt.axis('equal')
plt.title("Fraktal Sierpińskiego")
plt.show()
```

```python
# prompt: wykreśl zbiór Mandelbrodta z użyciem matplotlib

import matplotlib.pyplot as plt
import numpy as np

def mandelbrot(c, max_iter):
    z = 0
    n = 0
    while abs(z) <= 2 and n < max_iter:
        z = z*z + c
        n += 1
    return n

width, height = 500, 500
max_iter = 100

x_min, x_max = -2.0, 1.0
y_min, y_max = -1.5, 1.5

x = np.linspace(x_min, x_max, width)
y = np.linspace(y_min, y_max, height)

mandelbrot_set = np.zeros((height, width))

for i in range(height):
    for j in range(width):
        c = complex(x[j], y[i])
        mandelbrot_set[i, j] = mandelbrot(c, max_iter)

plt.figure(figsize=(8, 8))
plt.imshow(mandelbrot_set.T, extent=(x_min, x_max, y_min, y_max), cmap='hot')
plt.title("Zbiór Mandelbrota")
plt.xlabel("Re(c)")
plt.ylabel("Im(c)")
plt.show()
```
