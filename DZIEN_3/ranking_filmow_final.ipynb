```python
import pandas as pd
```

```python
import chardet

def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        raw_data = file.read()
        result = chardet.detect(raw_data)
        encoding = result['encoding']
        confidence = result['confidence']

        print(f"Wykryte kodowanie: {encoding}")
        print(f"Pewność: {confidence:.2%}")
        return encoding

# Przykład użycia
file_path = "mojedane/movies.dat"
detect_encoding(file_path)
```

```python
unames  = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('mojedane/users.dat', sep='::', engine='python', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('mojedane/ratings.dat', sep='::', engine='python', header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('mojedane/movies.dat', sep='::',encoding='ISO-8859-1', engine='python', header=None, names=mnames)
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings[:10]
```

```python
#ile ocen ma każdy z filmów
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:10]
```

```python
sorted_by_diff[::-1][:10]
```

```python
ratings_std_by_title = data.groupby('title')['rating'].std()
ratings_std_by_title = ratings_std_by_title.loc[active_titles]
ratings_std_by_title.sort_values(ascending=False)[:10]
```
