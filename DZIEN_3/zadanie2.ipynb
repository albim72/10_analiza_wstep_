```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generowanie danych za pomocą NumPy
data = np.random.randn(1000)

# Rysowanie histogramu z użyciem Matplotlib
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram z rozkładu normalnego")
plt.xlabel("Wartości")
plt.ylabel("Częstotliwość")
plt.show()
```

```python
# Tworzenie danych za pomocą Pandas i NumPy
data = pd.DataFrame({
    'Wzrost (cm)': np.random.normal(170, 10, 100),
    'Waga (kg)': np.random.normal(70, 15, 100)
})

# Rysowanie wykresu punktowego z użyciem Matplotlib
plt.scatter(data['Wzrost (cm)'], data['Waga (kg)'], color='green')
plt.title("Wykres punktowy wzrostu względem wagi")
plt.xlabel("Wzrost (cm)")
plt.ylabel("Waga (kg)")
plt.show()
```

```python
# Wczytanie danych z pliku CSV
df = pd.read_csv('dane.csv')

# Wyświetlenie pierwszych kilku wierszy danych
print(df.head())

# Sumaryczna sprzedaż dla każdego produktu
df['Łączna sprzedaż'] = df['Cena'] * df['Ilość sprzedana']
sprzedaz_sumaryczna = df.groupby('Produkt')['Łączna sprzedaż'].sum()

# Średnia cena dla każdego produktu
srednia_cena = df.groupby('Produkt')['Cena'].mean()

# Wyświetlenie wyników analizy
print("\nSumaryczna sprzedaż dla każdego produktu:")
print(sprzedaz_sumaryczna)
print("\nŚrednia cena dla każdego produktu:")
print(srednia_cena)

# Wykres słupkowy sumarycznej sprzedaży
sprzedaz_sumaryczna.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Sumaryczna sprzedaż dla każdego produktu')
plt.xlabel('Produkt')
plt.ylabel('Łączna sprzedaż (PLN)')
plt.show()
```
